```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
df = pd.read_csv("C:/Users/malvi/Desktop/data-hw-Malvika-R5/Titanic/dataset_201_guide.csv")
```

## 1. Who were the first-class passengers?
- **Requirements:**
  - Only display the name and class of the passengers

```python
is_first_class = df["Pclass"] == 1
columns_to_show = ["Name", "Pclass"]

df[is_first_class][columns_to_show]      # Inlining:--> df[df["Pclass"] == 1][["Name", "Pclass"]]
```

## 2. Who were the females in the first-class?
- **Requirements:**
  - Only display the name and gender of the passengers

```python
is_female = df["Sex"] == "female"
columns_to_show = ["Name", "Pclass", "Sex"]

df[is_female & is_first_class][columns_to_show]     
#Inlining:--> df[(df["Pclass"] == 1) & (df["Sex"] == "female")][["Name", "Pclass", "Sex"]]
```

## 3. Who were the first-class female passengers who were younger than 25 y/o?
- **Requirements:**
  - Display information in descending age
  - Only display the name, age, gender, and class of the passengers

```python
is_younger_than_25 = df["Age"] < 25
columns_to_show = ["Name", "Age", "Sex","Pclass"]

df[is_female & is_younger_than_25 & is_first_class][columns_to_show].sort_values(by="Age", ascending=False)
# inlining --> df[(df["Sex"]=="female") & (df["Pclass"] == 1) & (df["Age"]< 25)][["Name", "Age", "Sex","Pclass"]].sort_values(by="Age", ascending=False)
```
